**Ensemble methods** in machine learning refer to techniques that use **multiple learning algorithms** to obtain better predictive performance than could be obtained from any of the constituent learning algorithms alone. These methods combine the predictions from multiple models to improve the accuracy of the results.

**Ensemble learning methods** are made up of a set of classifiers—e.g. decision trees—and their predictions are aggregated to identify the most popular result. The most well-known ensemble methods are bagging, also known as bootstrap aggregation, and boosting. In 1996, Leo Breiman  introduced the bagging method; in this method, a random sample of data in a training set is selected with replacement—meaning that the individual data points can be chosen more than once. After several data samples are generated, these models are then trained independently, and depending on the type of task—i.e. regression or classification—the average or majority of those predictions yield a more accurate estimate. This approach is commonly used to reduce variance within a noisy dataset.

```python
from sklearn import datasets
from sklearn.ensemble import RandomForestClassifier
from sklearn import tree
from sklearn.model_selection import train_test_split as tts

from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = datasets.load_breast_cancer(as_frame=True)
```

```python
df = data['data']
df['target'] = data['target']
df.head()
```

### Simple Decision Tree

> The problem with decision tree is that they are **highly** sensitive to the training data which could result in high variance so model might fail to generalize. The slightest change to a value in the training data could result in a completely different tree!

```python
X = df.iloc[:, :-1]
y = df['target']

X_train, X_test, y_train, y_test = tts(X, y, random_state=42)
```

```python
dt_model = tree.DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train, y_train)

plt.figure(figsize=(20,20))
tree.plot_tree(dt_model, filled=True, feature_names = list(df.columns[:-1]));
```

### Random Forest Algorithm

combines the outputs of multiple decision trees to make a final prediction. Each tree in the ensemble is built using a data sample drawn from the training set, and the algorithm is used for both classification and regression problems. The random forest algorithm is known for its ease of use and flexibility, and it is widely used in various fields such as finance, medicine, and e-commerce.

**Algorithm**
1. Draw a sample dataset from the original dataset (with replacement, meaning a single instance can be drawn multiple times also called as Bootstrapping)
2. A decision tree is trained on each bootstrapped subset independently. (Random independent features will be used as the predictor variable for each independent decision tree model)
3. Finally for the prediction, each trained tree model will be utilized and the majority votes will be used as the final prediction. (Combining of result is also called as aggregation).

Random Forest combines two methods bootstrapping and aggregating, this procedure is called **Bagging.**

```python
train_set = pd.concat([X_train, y_train], axis=1)
```

```python
n_feature = 10
forest = []
for i in range(10):
    subset = train_set.sample(n=50, replace=True)
    subset_x, subset_y = subset.iloc[:, :-1], subset['target']
    sub_features = np.random.choice(subset.columns[:-1], size=n_feature)
    sub_dt = tree.DecisionTreeClassifier(random_state=42).fit(subset[sub_features], subset_y)
    forest.append(sub_dt)
```

```python
forest[0].predict(X_test[forest[0].feature_names_in_])
```
