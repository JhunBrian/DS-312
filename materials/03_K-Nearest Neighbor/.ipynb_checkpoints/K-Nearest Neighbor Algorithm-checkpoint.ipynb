```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn import neighbors
from sklearn.preprocessing import LabelEncoder
```

### Load Dataset

```python
df = sns.load_dataset('iris')
df = df.sample(n=len(df), random_state=42).reset_index(drop=True)
df.head()
```

```python
plt.title("Iris Dataset")
sns.scatterplot(data=df, x='petal_length', y='petal_width', hue='species');
plt.grid()
```

```python
train_size, test_size = 0.70, 0.30
```

```python
train_data = df.iloc[:int(len(df) * train_size), :]
test_data = df.iloc[int(len(df) * train_size):, :]

X_train, y_train = train_data.iloc[:, :-1], train_data['species']
X_test, y_test = test_data.iloc[:, :-1], test_data['species']
```

**The Euclidean Disntance**

\begin{equation}
    dist(a, b) = \sqrt{\sum_{i=1}^n (a_i - b_i)^2}
\end{equation}

Where:
- $a$ is the unseen data point.
- $b$ is the seen data point.
- $n$ is the number of samples.

**Utility Function: Euclidean Distance**

```python
def euclidean_distance(a_series, b_series):
    sums = 0
    for a, b in zip(a_series, b_series):
        sums += (a - b)**2
    return np.sqrt(sums)
```

\begin{equation}
    \textit{unseen data} = [4.6, 3.4, 1.4, 0.3, ?]
\end{equation}

```python
for idx, row in X_train.iterrows():
    print(row)
    break
```

```python
# Try changing the k and the index

k = 50
index = 34
distance = [euclidean_distance(X_test.iloc[index], row) for idx, row in X_train.iterrows()]

X_copy = train_data.copy()
X_copy['distance'] = distance

sorted = X_copy.sort_values(by='distance')
sorted_df = sorted.head(k)

counts = sorted_df['species'].value_counts()
proportion = counts / counts.sum()


fig, ax = plt.subplots(1, 2, figsize=(10,4))

sns.barplot(x=proportion.index, y=proportion, ax=ax[0])

sns.scatterplot(data=X_copy, x='petal_length', y='petal_width', hue='species', ax=ax[1])
ax[1].scatter(x=X_test.iloc[index]['petal_length'], y=X_test.iloc[index]['petal_width'], c='gray', label='unseen')
ax[1].legend()

print(f"True Value: {y_test.iloc[index]}")
print(f"Predicted Value: {proportion.index[0]}")
```

## Implementing KNN Classifier with Sklearn

```python
clf = neighbors.KNeighborsClassifier(n_neighbors=k).fit(X_train, y_train)

clf.predict(X_test)
```
